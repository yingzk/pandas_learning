# Iris纸鸢花数据 - 删除数据

![](images/10.jpg)

### 导入库

```python
import numpy as np
import pandas as pd
```

### 加载数据

```python
iris = pd.read_csv('data/iris.csv')
iris.head(3)
```

### 创建数据框的列名称

```python
iris.columns = ['sepal_length','sepal_width', 'petal_length', 'petal_width', 'class']
iris.head(3)
```

### 数据框中有缺失值吗？

```python
pd.isnull(iris).sum()
```

### 将列petal_length的第10到19行设置为缺失值

```python
iris.iloc[10:20,2:3] = np.nan
iris.head(20)
```

### 将缺失值全部替换为1.0

```python
iris.petal_length.fillna(1, inplace=True)
```

```python
iris
```

### 删除列class

```python
del iris['class']
iris.head(3)
```

### 将数据框前三行设置为缺失值

```python
iris.iloc[0:3, :] = np.nan
iris.head()
```

### 删除有缺失值的行

```python
iris = iris.dropna(how='any')
iris
```

### 重新设置索引

```python
iris = iris.reset_index(drop = True)
iris.head()
```
